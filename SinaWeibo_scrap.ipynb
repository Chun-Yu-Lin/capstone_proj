```python
import requests
import pandas as pd

articles = []
for i in range(5):
    res = requests.get('https://m.weibo.cn/api/container/getIndex?containerid=231522type%3D1%26t%3D10%26q%3D%23%E6%96%B0%E5%86%A0%E8%82%BA%E7%82%8E%23&extparam=%23%E6%96%B0%E5%86%A0%E8%82%BA%E7%82%8E%23&luicode=10000011&lfid=100103type%3D64%26q%3D%23%E6%96%B0%E5%86%A0%E8%82%BA%E7%82%8E%23%26t%3D0&page_type=searchall&page={}'.format(i))
    jd = res.json()
    articles.extend([rec['mblog'] for rec in jd['data']['cards'] if rec.get('mblog')])
    
df = pd.DataFrame(articles)
```

```python
from bs4 import BeautifulSoup

def parseArticle(e):
    soup = BeautifulSoup(e, 'lxml')
    return soup.text

df['text'].map(parseArticle)
```
